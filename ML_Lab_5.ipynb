```python
sex = pd.get_dummies(data['Gender'])
data.drop(["Gender"], axis = 1, inplace = True)
data.head()
```

```python
data = pd.concat([sex,data],axis = 1)
data.head()
```

```python
age = []
AGE = data["Age"].to_numpy()     
for i in AGE:
    age.append(i)

sal = []
Sal = data["EstimatedSalary"].to_numpy()
for i in Sal:
    sal.append(i)

def normalize(data):
    normalized = []
    range = max(data) - min(data)
    minn = min(data)
    for j in data:
        normalized.append((j - minn)/range)
    return normalized

norm_age = normalize(age)
norm_sal = normalize(sal)
norm_age = np.array(norm_age)
norm_sal = np.array(norm_sal)
```

```python
def standard(data,mean,std):
    standerdized = []
    for j in data:
        standerdized.append((j-mean)/std)
    return standerdized
std_age = standard(age,data["Age"].mean(),data["Age"].std())
std_sal = standard(sal,data["EstimatedSalary"].mean(),data["EstimatedSalary"].std())

std_sal = np.array(std_sal)
std_age = np.array(std_age)
```

```python
stdage = pd.DataFrame(std_age,columns = ['Std Age'])
stdsal = pd.DataFrame(std_sal,columns = ['Std Salary'])
normage = pd.DataFrame(norm_age,columns = ['Norm Age'])
normsal = pd.DataFrame(norm_sal,columns = ['Norm Sal'])
```

```python
data.drop(["Age","EstimatedSalary"], axis = 1, inplace = True)
data = pd.concat([stdage,stdsal,data],axis = 1)
data
```
